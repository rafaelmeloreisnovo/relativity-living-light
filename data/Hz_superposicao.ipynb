# Hz Superposição - Análise da Taxa de Expansão de Hubble

## Propósito Científico
Este notebook implementa e visualiza a equação de Friedmann modificada com superposição fotônica.
Calcula H(z) para diferentes modelos cosmológicos e compara com o ΛCDM padrão.

## Interpretação Simbiótica
A superposição fotônica representa a dualidade luz-matéria, transitando suavemente entre
comportamento expansivo (energia escura) e atrativo (matéria escura). Este código materializa
matematicamente o conceito de 'luz viva' que assume estados diferentes conforme a evolução cósmica.

## Equação Principal
```
E²(a) = Ω_r a⁻⁴ + Ω_m a⁻³ + Ω_Λ + Ω_s0[f(a) + (1-f)a⁻³]
onde f(z) = 1 / (1 + exp((z - z_t)/w_t))
```

## Parâmetros
- `H0`: Constante de Hubble hoje (km/s/Mpc)
- `Omega_m`: Densidade de matéria
- `Omega_L`: Constante cosmológica
- `Omega_s0`: Densidade de superposição fotônica
- `z_t`: Redshift de transição
- `w_t`: Largura da transição


# H(z) com termo de **ρ_superposição**

Este notebook compara o modelo **ΛCDM** com um modelo estendido que inclui um termo adicional de densidade
associado à **superposição fotônica**:

\[ H(z)^2 = H_0^2 \left[ \Omega_m (1+z)^3 + \Omega_r (1+z)^4 + \Omega_\Lambda + \Omega_{\text{sup}} a^{-n} \right] \]

Ajuste `Omega_sup` e `n_sup` abaixo para explorar diferentes cenários.

````
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros básicos (ajuste aqui)
H0 = 70.0       # km/s/Mpc
Omega_m = 0.3
Omega_r = 1e-4
Omega_L = 0.7

# Termo de superposição fotônica (ajuste aqui)
Omega_sup = 0.1   # fração de densidade hoje
n_sup = 2.0       # evolução: rho_sup ~ a^{-n_sup}

# Grade de redshift
z = np.linspace(0, 3, 300)
a = 1.0 / (1.0 + z)

# Modelos
Hz_LCDM = H0 * np.sqrt(Omega_m * (1+z)**3 + Omega_r * (1+z)**4 + Omega_L)
Hz_sup  = H0 * np.sqrt(Omega_m * (1+z)**3 + Omega_r * (1+z)**4 + Omega_L + Omega_sup * a**(-n_sup))

# Plot
plt.figure(figsize=(7,5))
plt.plot(z, Hz_LCDM, label="ΛCDM (padrão)")
plt.plot(z, Hz_sup, label=f"Com ρ_superposição (Ω_sup={Omega_sup}, n={n_sup})", linestyle="--")
plt.xlabel("Redshift z")
plt.ylabel("H(z) [km/s/Mpc]")
plt.title("Comparação H(z): ΛCDM vs. Modelo com Superposição Fotônica")
plt.legend()
plt.grid(True)
plt.show()
````
